```python
#1. Usar el método corr() para ver qué variables tienen mayor correlación con el precio
correlacion_precio = df.corr(numeric_only=True)['price'].sort_values(ascending=False)
correlacion_precio
```

```python
# Seleccionar 3 variables con alta correlación
variables_predictoras = ['engine-size', 'curb-weight', 'horsepower']
X = df[variables_predictoras]
y = df['price']
```

```python
# 2. Realizar el entrenamiento (método fit)

lm3 = LinearRegression()
lm3.fit(X, y)
```

```python
# 3. Realizar la predicción (método predict) y mostrar todos los valores
predicciones = lm3.predict(X)
print(predicciones[:10])
```

```python
# 4. Mostrar el valor de la intercepción (Intercept_)
print(f"Intercept: {lm3.intercept_}")
```

```python
# 5. Mostrar el valor de los coeficientes (coef_)
for variable, coef in zip(variables_predictoras, lm3.coef_):
    print(f"{variable}: {coef}")
```

```python
# 6. Realizar una predicción manualmente con la fórmula
# Tomamos el primer ejemplo del dataset
ejemplo = X.iloc[0]
print(f"\nValores del ejemplo:")
for variable in variables_predictoras:
    print(f"{variable}: {ejemplo[variable]}")

prediccion_manual = lm3.intercept_
for i, variable in enumerate(variables_predictoras):
    prediccion_manual += lm3.coef_[i] * ejemplo[variable]

print(f"\nPredicción manual: ${prediccion_manual:.2f}")
```

```python
# 7. Realizar una predicción con el método predict

prediccion_metodo = lm3.predict([ejemplo.values])[0]
print(f"Predicción con predict: ${prediccion_metodo:.2f}")
print(f"Diferencia: {abs(prediccion_manual - prediccion_metodo):.10f}")
```

```python
# 8. Mostrar el coeficiente de determinación (método score)

r2_score = lm3.score(X, y)
print(f"R² Score: {r2_score:.4f}")
print(f"El modelo explica el {r2_score*100:.2f}% de la variabilidad del precio")
```
